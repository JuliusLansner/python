```python
#I want to clean up the date abit, seeing as 20141013T000000 is a weird format.
df['date'] = pd.to_datetime(df['date']) #corretly set it to a humanly readable datetime.
```

```python
#bathrooms and floors are weird. 2.25 bathrooms?!
df['bathrooms'] = df['bathrooms'].astype(int)#setting bathrooms to an int, as opposed to a float/double.
df['floors'] = df['floors'].astype(int)#setting floors to an int(whole number), as opposed to a float/double.
```

```python
#shaving some potentially useless information off
df.drop(['id','lat','long'],axis = 1)# drops columns, givern by axis=1. axis=1 refering to dropping colums, in the drop function.
```
